```python
 



df = result_df


# Define a function to check if a value starts with a specific string 

def check_startswith(value, prefix): 

    if pd.isna(value): 

        return False 

    return str(value).startswith(prefix) 

def indent_row(row):

    sorted_row = row.values[np.argsort(row.isna())]

    return pd.Series(sorted_row,index=row.index)
  

# Replace values in "BioProject", "BioSample", and "Assembly" columns that do not start with the respective strings 

for col in ["BioProject", "BioSample", "Assembly"]: 

    df[f'{col}_invalid'] = df[col].apply(lambda x: x if not check_startswith(x, col) else np.nan) 

    df[col] = df[col].apply(lambda x: x if check_startswith(x, col) else np.nan) 

  



# Move invalid data to the end of the dataframe 

invalid_columns = [col for col in df.columns.astype(str) if col.endswith('_invalid')]  # Convert column names to strings 

df_invalid = df[invalid_columns] 

df.drop(columns=invalid_columns, inplace=True) 

df = pd.concat([df, df_invalid], axis=1) 

  

# Rename columns after "Assembly" 

assembly_index = df.columns.get_loc("Assembly") 

new_column_names = {df.columns[i]: f"Reference_{i-assembly_index}" for i in range(assembly_index + 1, len(df.columns))} 

df.rename(columns=new_column_names, inplace=True) 

column_names = df.columns.tolist()

# Indentation of references

get_assembly_index = df.columns.get_loc("Assembly")

df.iloc[:,get_assembly_index+1:] = df.iloc[:,get_assembly_index+1:].apply(indent_row, axis=1)

df.columns = column_names

#

df['BioSample'] = df.apply(lambda row: row['Reference_2'] if check_startswith(row['Reference_2'], 'BioSample:') else row['BioSample'], axis=1) 

df['Reference_2'] = df.apply(lambda row: np.nan if check_startswith(row['Reference_2'], 'BioSample:') else row['Reference_2'], axis=1) 

df['Assembly'] = df.apply(lambda row: row['Reference_1'] if check_startswith(row['Reference_1'], 'Assembly:') else row['Assembly'], axis=1) 

df['Reference_1'] = df.apply(lambda row: np.nan if check_startswith(row['Reference_1'], 'Assembly:') else row['Reference_1'], axis=1) 

 
# Indentation of references

get_assembly_index = df.columns.get_loc("Assembly")

df.iloc[:,get_assembly_index+1:] = df.iloc[:,get_assembly_index+1:].apply(indent_row, axis=1)

df.columns = column_names


# Display the first few rows of the modified DataFrame 

df.head() 

 
```
